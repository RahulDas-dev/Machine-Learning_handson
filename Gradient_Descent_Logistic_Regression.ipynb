
# Logistic Regression With Gradient Descent



![log_reg](./logistic_regression.svg)


$$\large Z = \sum \limits _{i=1} ^{n} w_{i}\times x_{i} + b \Rightarrow W^{T} X + b  \quad\quad\quad\quad  \forall W\in \mathbb{R}^{n}$$

$$\large \hat{Y} = \sigma(z) \Rightarrow\sigma( W^{T} X + b) $$


$$\large \hat{Y} = f(x;w_{1},w_{2},w_{3},\dots,w_{n},b) = f(x;\theta )  \quad\text{where} \quad \theta \equiv {w_{1},w_{2},w_{3},\dots,w_{n},b}$$ 



#### Cost Function

$$\large L_{CE}(\hat{Y},Y) = -\frac{1}{m} \sum_{i=1}^{m} y_{i}.log(\hat{y}_{i}) + (1-y_{i}).log(1-\hat{y}_{i})$$

#### Gradient Descent  

$$\large \theta_{t+1} = \theta_{t} - \eta \times \nabla_{\theta} L_{CE}(\hat{Y},Y)   $$

$$ \large \updownarrow $$

$$\large \theta_{t+1} = \theta_{t} - \eta \times \nabla_{\theta} L_{CE}(f(x;\theta),Y)   $$




#### Claculations of Gredient 

$$\nabla_{\theta} L_{CE}(f(x;\theta),Y) \Rightarrow \nabla_{\theta} L_{CE}(\hat{Y},Y) = \begin{bmatrix}
\frac{\partial L_{CE}(f(x;\theta),Y) }{\partial w_{1}} \\
\frac{\partial L_{CE}(f(x;\theta),Y) }{\partial w_{2}} \\
\frac{\partial L_{CE}(f(x;\theta),Y) }{\partial w_{2}} \\
\vdots\\
\frac{\partial L_{CE}(f(x;\theta),Y) }{\partial w_{n}} \\
\frac{\partial L_{CE}(f(x;\theta),Y) }{\partial b} \\
\end{bmatrix}  
$$


#### Partial Diff of Cost Function with respect to Weights

$\large \frac{\partial L_{CE}(f(x;\theta),Y) }{\partial w_{i}} = \frac{\partial L_{CE}(\hat{Y},Y)} {\partial w_{i} } $

$\large = \frac{\partial  }{\partial w_{i}}\left[ -\frac{1}{m} \sum_{k=1}^{m} y^{k}.log(\hat{y}^{k}) + (1-y^{k}).log(1-\hat{y}^{k}) \right]$

$\large = - \frac{1}{m} \sum_{k=1}^{m}  y^{k} \frac{\partial log(\hat{y}^{k}) }{\partial w_{i}}  + (1-y^{k}) \frac{\partial log(1-\hat{y}^{k}) }{\partial w_{i}} $

$\large = - \frac{1}{m} \sum_{k=1}^{m}  y^{k} \frac{\partial log(\hat{y}^{k}) }{\partial \hat{y}^{k}} \frac{\partial \hat{y}^{k}}{\partial w_{i}}  + (1-y^{k}) \frac{\partial log(1-\hat{y}^{k}) }{\partial \hat{y}^{k}} \frac{\partial \hat{y}^{k}}{\partial w_{i}} $


$\large = - \frac{1}{m} \sum_{k=1}^{m}  \frac{y^{k}}{\hat{y}^{k}} \frac{\partial \hat{y}^{k}}{\partial w_{i}}  +  \frac{-\hat{y}^{k}(1-y^{k})  }{1-\hat{y}^{k}} \frac{\partial \hat{y}^{k}}{\partial w_{i}} $

$\large = - \frac{1}{m} \sum_{k=1}^{m} \left( \frac{y^{k}-\hat{y}^{k}} { \hat{y}^{k} (1-\hat{y}^{k})} \right)\frac{\partial \hat{y}^{k}}{\partial w_{i}} $


#### Partial Differentiation of yhat with respect to weights

$\large \frac{\partial \hat{y}^{k} }{\partial w_{i}} = \frac{\partial \sigma(z^{k}) }{\partial w_{i}} =  \frac{\partial \sigma(z^{k}) }{\partial z^{k}} \frac{\partial z^{k}}{\partial w_{i}} =  \frac{\partial \sigma(z^{k}) }{\partial z^{k}}  \frac{\partial \sum_{i=1}^{n} w_{i} x^{k}_{i} + b }{\partial w_{i}}= \frac{\partial \sigma(z^{k}) }{\partial z^{k}} x^{k}_{i} = \sigma(z^{k})(1 - \sigma(z^{k}) ) x^{k}_{i} = \hat{y^{k}}(1 - \hat{y}^{k} ) x^{k}_{i}$



$$\frac{\partial \hat{y}^{k} }{\partial w_{i}} = \hat{y}^{k}  (1 - \hat{y}^{k} )  x^{k}_{i}$$

#### Replaceing equation 5 


$\large \frac{\partial L_{CE}(\hat{y},y) }{\partial w_{i}} = - \frac{1}{m} \sum_{k=1}^{m} \left( \frac{y^{k}-\hat{y}^{k}} { \hat{y}^{k}  (1-\hat{y}^{k})} \right) \frac{\partial \hat{y}^{k}}{\partial w_{i}} = - \frac{1}{m} \sum_{k=1}^{m} \left( \frac{y^{k}-\hat{y}^{k}} { \hat{y}^{k} (1-\hat{y}^{k})} \right)\hat{y}^{k} (1 - \hat{y}^{k} ) x^{k}_{i}$

$$\frac{\partial L_{CE}(\hat{y},y) }{\partial w_{i}} = - \frac{1}{m} \sum_{k=1}^{m}  \left(y^{k}-\hat{y}^{k}\right)  x^{k}_{i}$$


$$\frac{\partial L_{CE}(\hat{y},y) }{\partial b} = - \frac{1}{m} \sum_{k=1}^{m}  \left(y^{k}-\hat{y}^{k}\right) $$


#### Expressing gredient interms of matix


$\large \frac{\partial L_{CE}(\hat{Y},Y) }{\partial w_{i}} = - \frac{1}{m} \sum_{k=1}^{m}  \left(y^{k}-\hat{y}^{k}\right) x^{k}_{i}$


$\large = - \frac{1}{m} \left[\left(y^{1}-\hat{y}^{1}\right)  x^{1}_{i} +\left(y^{2}-\hat{y}^{2}\right)  x^{2}_{i} + \left(y^{3}-\hat{y}^{3}\right)  x^{3}_{i} + \dots + \left(y^{m}-\hat{y}^{m}\right)  x^{m}_{i} \right ] $



$\large =  - \frac{1}{m} \begin{bmatrix} x^{1}_{i}&x^{2}_{i}&x^{3}_{i}&\dots&x^{m}_{i} \end{bmatrix} \begin{bmatrix}
y^{1}-\hat{y}^{1} \\
y^{2}-\hat{y}^{2} \\
y^{3}-\hat{y}^{3} \\
\vdots\\
y^{m}-\hat{y}^{m}\\
\end{bmatrix} 
$

$$A =  \begin{bmatrix} 
x^{1}_{1} & x^{1}_{2} & x^{1}_{3} & \dots & x^{1}_{i} & \dots & x^{1}_{n} \\
x^{2}_{1} & x^{2}_{2} & x^{2}_{3} & \dots & x^{2}_{i} & \dots & x^{2}_{n} \\
x^{3}_{1} & x^{3}_{2} & x^{3}_{3} & \dots & x^{3}_{i} & \dots & x^{3}_{n} \\
\vdots & & \dots & & \vdots\\
x^{p}_{1} & x^{p}_{2} & x^{p}_{3} & \dots & x^{p}_{i} & \dots & x^{i}_{n} \\
\vdots & & \dots & & \vdots\\
x^{m}_{1} & x^{m}_{2} & x^{m}_{3} & \dots & x^{m}_{i} & \dots & x^{m}_{n} \\
\end{bmatrix}$$

$$\frac{\partial L_{CE}(\hat{Y},Y) }{\partial w_{i}} = - \frac{1}{m} A_{i^{th} Column } \times \begin{bmatrix}
y^{1}-\hat{y}^{1} \\
y^{2}-\hat{y}^{2} \\
y^{3}-\hat{y}^{3} \\
\vdots\\
y^{m}-\hat{y}^{m}\\
\end{bmatrix} 
$$


$$\frac{\partial L_{CE}(\hat{Y},Y) }{\partial w_{i}} = - \frac{1}{m} A^{T}_{i^{th} row } \times \begin{bmatrix}
y^{1}-\hat{y}^{1} \\
y^{2}-\hat{y}^{2} \\
y^{3}-\hat{y}^{3} \\
\vdots\\
y^{m}-\hat{y}^{m}\\
\end{bmatrix} 
$$

#### Finally Gredient in Matrix format


$$\nabla_{\theta} L_{CE}(\hat{Y},Y) = \begin{bmatrix}
\frac{\partial L_{CE}(\hat{y},y) }{\partial w_{1}} \\
\frac{\partial L_{CE}(\hat{y},y) }{\partial w_{2}} \\
\frac{\partial L_{CE}(\hat{y},y) }{\partial w_{2}} \\
\vdots\\
\frac{\partial L_{CE}(\hat{y},y) }{\partial w_{n}} \\
\frac{\partial L_{CE}(\hat{y},y) }{\partial b} \\
\end{bmatrix}  
= 
\begin{bmatrix}
\nabla_{\theta} L_{CE}(\hat{y},y)_{w}\\
\frac{\partial L_{CE}(\hat{y},y) }{\partial b} \\
\end{bmatrix}
$$


$$ \nabla_{\theta} L_{CE}(\hat{Y},Y)  = \begin{bmatrix}
\frac{\partial L_{CE}(\hat{y},y) }{\partial w_{1}} \\
\frac{\partial L_{CE}(\hat{y},y) }{\partial w_{2}} \\
\frac{\partial L_{CE}(\hat{y},y) }{\partial w_{2}} \\
\vdots\\
\frac{\partial L_{CE}(\hat{y},y) }{\partial w_{n}} \\
\end{bmatrix} 
=
- \frac{1}{m} A^{T} \times \begin{bmatrix}
y^{1}-\hat{y}^{1} \\
y^{2}-\hat{y}^{2} \\
y^{3}-\hat{y}^{3} \\
\vdots\\
y^{m}-\hat{y}^{m}\\
\end{bmatrix}  ;  
\frac{\partial L_{CE}(\hat{y},y) }{\partial b} = - \frac{1}{m} \sum_{k=1}^{m}  \left(y^{k}-\hat{y}^{k}\right)
$$

### Gradient Descent Logistic Regression

```python
class GradientDescentLogisticRegression:
    
    def __init__(self, iteration=500, learning_rate=0.001):
        self.__iteration = iteration
        self.__learning_rate = learning_rate
        self.__epsilon = 1e-5
        self.__wights = None
        self.__bias = 0
      
    def get_weights(self):
        return self.__wights
    
    def get_bias(self):
        return self.__bias
    
    def fit(self,x,y):
        num_of_observation, num_of_feature = x.shape # Training Size Capture
        self.__wights = np.random.normal(loc=0,scale=1/num_of_feature, size=num_of_feature )
        #self.__wights = np.zeros(shape=num_of_feature ) # Weights initialization
        self.__bias = 0                                  # bias initialization  
        i ,history = 0, []
        while i < self.__iteration:                #    Training Loop Strt Here ##########
            yhat = self.__forward_propagation(x)   #    Forward Propagation     ##########
            grd_w, grd_b = self.__compute_grediant(x,y,yhat) # Gredient Calculation   ####
            self.__update_weights(grd_w, grd_b)    #    Weights and bais Update    #######   
            loss = self.__compute_loss(yhat,y)     #    Loss Computation    ##############
            TP,TN,FP,FN,precision,recall,F1 = self.__compute_accuracy(yhat,y) #  Accuracy Computation  #######
            history.append({                       #   Training History ##################
                            'iteration':i,
                            'loss': loss,
                            'TP':TP,'TN':TN,'FP':FP,'FN':FN,
                            'precision':precision,'recall':recall,'F1':F1
                           })
            if loss <= self.__epsilon:              # End Traing if Loss bellow Threshold
                break
            i += 1                                  # Traing Loop End Here
        print(f'Iteration {self.__iteration} Training Loss: {loss} TP:{TP},TN:{TN},FP:{FP},FN:{FN}')    
        return history
    
    def __forward_propagation(self,x):
        z = x.dot(self.__wights) + self.__bias
        return 1/( 1 + np.exp(-z) )
    
    def __compute_grediant(self, x,y,yhat):
        error_vector = yhat - y    
        num_of_observation, _ = x.shape
        grad_w = np.matmul(np.transpose(x),error_vector)/num_of_observation
        grad_b = np.mean(error_vector)
        return grad_w, grad_b
    
    def __update_weights(self, grd_w, grd_b):
        self.__wights = self.__wights - self.__learning_rate*grd_w
        self.__bias = self.__bias - self.__learning_rate*grd_b
    
    def __compute_loss(self,yhat,y):
        loss_1 = -1*y*np.log(yhat+self.__epsilon)
        loss_0 = -1*(1-y)*np.log(1-yhat+self.__epsilon)
        loss = np.mean(loss_1+loss_0)
        return loss
    
    def __compute_accuracy(self,yhat,y):
        y_pred = np.where(yhat >=.5,1,0)
        TP = np.sum(np.logical_and(y==1,y_pred==1))
        TN = np.sum(np.logical_and(y==0,y_pred==0))
        FP = np.sum(np.logical_and(y==0,y_pred==1))
        FN = np.sum(np.logical_and(y==1,y_pred==0))
        precision = TP/(TP+FP)  # Ratio of TP and positive predictions
        recall = TP/(TP+FN)     # Ratio of TP and Actual positive
        F1 = 2*(precision*recall)/(precision+recall)
        return TP,TN,FP,FN,precision,recall,F1
        
    def predict(self,x):
        y_pred =  self.__forward_propagation(x)
        return np.where(y_pred >=.5,1,0)
```

### Fitting Model to Data

```python
cls = GradientDescentLogisticRegression(iteration=1400, learning_rate=0.08)
history = cls.fit(X_data, Y_data)
y_pred = cls.predict(X_data)
print(
    f'X_data shape {X_data.shape},Y_data shape {Y_data.shape} Y hat shape {y_pred.shape}')
print(np.unique(Y_data, return_counts=True))
print(np.unique(y_pred, return_counts=True))
```

```python
from itertools import product

def confusion_matrix(y_true,y_pred):
    unique_cls = np.unique(np.concatenate((y_true,y_pred)))   # Collecting unique class
    n_cls = unique_cls.shape[0]                               # Counting number of unique class
    zip_data = list(zip(y_true,y_pred))                       # zipping true and prediction class
    result=[]                              
    for c in product(unique_cls,repeat=2):                    # Finding out the NC2 on unique classes
        result.append(zip_data.count(c))                      # Counting   
    return np.asarray(result).reshape(n_cls,n_cls)

confu_mat = confusion_matrix(Y_data,y_pred)
print(confu_mat)
```

```python
def plot_confusion_matrix(confusion_matix):
    row,col = confusion_matix.shape
    fig, ax = plt.subplots(figsize=(1*row, 1*col))
    ax.imshow(confusion_matix, cmap=plt.cm.Blues, alpha=0.3)
    for r,c in product(range(row),range(col)):
        ax.text(x=c, y=r, s=confusion_matix[r, c],
                    va='center', ha='center', size='x-large')
    
    ax.set_xlabel('Predictions', fontsize=15)
    ax.set_ylabel('Actuals', fontsize=15)
    #ax.set_xticks(list(range(c+1)))
    ax.set_yticks(list(range(r+1)))

    ax.set_title('Confusion Matrix', fontsize=20)
    plt.show()

plot_confusion_matrix(confu_mat)    
```

### Plotting Decission Boundary

$z = \sum \limits _{i=1} ^{n} w_{i}\times x_{i} + b \Rightarrow w_{1} x_{1} + w_{2} x_{2}+ b$

$\hat{y} = \sigma(z) \Rightarrow\sigma( w_{1} x_{1} + w_{2} x_{2}+ b) $

at boundary $\hat{y} = 0.5 \Rightarrow \sigma(z) = 0.5$ 

$\frac{1}{1 + exp^{-z}} = 0.5$ 

$1 + exp^{-z} = 2 \Rightarrow exp^{-z} = 1$ 

$z = 0 \Rightarrow w_{1} x_{1} + w_{2} x_{2}+ b = 0$ 

so decession boundary is represented by  $w_{1} x_{1} + w_{2} x_{2}+ b = 0$  this line 
slope of the is $m = -\frac{w_{1}}{w_{2}}$ and intercept is $ c = - \frac{b}{w_{2}}$
